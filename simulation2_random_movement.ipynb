```
import random
import folium

enlem = []
boylam = []
usage1 = []
usage2 = []
m = folium.Map(location=[41.104385,29.022716] ,zoom_start =20)
nodes = int(input("Please enter total number of nodes = "))
apps = int(input("Please enter number of app usage = "))

for i in range (0,nodes):

  x = random.uniform(41.1045, 41.1029)
  enlem.append(x)
  y = random.uniform(29.0239,29.0217)
  boylam.append(y)
  folium.Marker(location=[x,y],tooltip="People that is not using App").add_to(m)

folium.Marker(location=[41.103876,29.022749],tooltip="Gateway",icon=folium.Icon(color='green')).add_to(m)

for j in range (0,apps):
  blue1 = random.choice(enlem)
  usage1.append(blue1)
  blue2 = random.choice(boylam)  
  usage2.append(blue2)
  folium.Marker(location=[blue1,blue2],tooltip=j,icon=folium.Icon(color='red')).add_to(m)
  folium.CircleMarker(
      location=[blue1,blue2],
      radius=10,
      color='red',
      fill=True,
      fill_color='red'
  ).add_to(m)


m
```

```
a = [1,2,3,4]
step1_enlem = []   #enlem
step1_boylam = []   #boylam

def move_north(i):
  k = float(usage1[i] + 0.00005968)
  step1_enlem.append(k)
  l = usage2[i]
  step1_boylam.append(l)
  folium.Marker(location=[k,usage2[i]],tooltip=i,icon=folium.Icon(color="orange")).add_to(m)

def move_south(i):
  k = float(usage1[i] - 0.00005968)
  step1_enlem.append(k)
  l = usage2[i]
  step1_boylam.append(l)
  folium.Marker(location=[k,usage2[i]],tooltip=i,icon=folium.Icon(color="orange")).add_to(m)

def move_west(i):
  k = float(usage2[i] + 0.00005968)
  step1_boylam.append(k)
  l = usage1[i]
  step1_enlem.append(l)
  folium.Marker(location=[usage1[i],k],tooltip=i,icon=folium.Icon(color="orange")).add_to(m)

def move_east(i):
  k = float(usage2[i] - 0.00005968)
  step1_boylam.append(k)
  l = usage1[i]
  step1_enlem.append(l)
  folium.Marker(location=[usage1[i],k],tooltip=i,icon=folium.Icon(color="orange")).add_to(m)

for i in range (apps):
  switch = random.choice(a)
  if switch == 1:
    move_north(i)
  elif switch == 2:
    move_south(i)
  elif switch == 3:
    move_west(i)
  elif switch == 4:
    move_east(i)

m
```

```
from math import sin, cos, sqrt, atan2, radians


def calculate(lat1,lon1,lat2,lon2):
  # approximate radius of earth in km
  R = 6373.0

  dlon = lon1 - lon2
  dlat = lat1 - lat2

  a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
  c = 2 * atan2(sqrt(a), sqrt(1 - a))

  distance = R * c
  metre = distance*1000
  if metre <= 10:
    print(k,"Result:", distance)
    print(k,"Should be:", metre , "m")

def calculate_gateway(lat3,lon3,lat2,lon2):
  # approximate radius of earth in km
  R = 6373.0

  dlon = lon3 - lon2
  dlat = lat3 - lat2

  a = sin(dlat / 2)**2 + cos(lat3) * cos(lat2) * sin(dlon / 2)**2
  c = 2 * atan2(sqrt(a), sqrt(1 - a))

  distance = R * c
  metre = distance*1000
  
  if metre <= 10:
    print(k,"Connected gateway:", metre , "m")

  

for k in range(len(step1_enlem)):
  lat1 = radians(step1_enlem[k])
  lon1 = radians(step1_boylam[k])
  lat3 = radians(41.103876)
  lon3 = radians(29.022749)
  for i in range(k):
    lat2 = radians(step1_enlem[i])
    lon2 = radians(step1_boylam[i])
    calculate(lat1,lon1,lat2,lon2)
    calculate_gateway(lat3,lon3,lat2,lon2)

```
